```python
# Step 3: Apply Band-Pass Filter and Handle Missing Values

from scipy.signal import butter, filtfilt
import pandas as pd

# Function to create a band-pass filter
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Sampling frequency and filter parameters
fs = 256  # Sampling frequency, adjust based on your actual dataset
lowcut = 0.5  # Low cut frequency for band-pass filter
highcut = 50  # High cut frequency for band-pass filter

# Apply band-pass filter to EEG channels like 'Alpha_TP9', 'Theta_AF7', etc.
dfs[0]['filtered_Alpha_TP9'] = bandpass_filter(dfs[0]['Alpha_TP9'], lowcut, highcut, fs)
dfs[0]['filtered_Theta_AF7'] = bandpass_filter(dfs[0]['Theta_AF7'], lowcut, highcut, fs)

# Handle missing data by forward-filling missing values
dfs[0].fillna(method='ffill', inplace=True)

# Check if there are any remaining missing values
print("Missing values in each column after filling:")
print(dfs[0].isnull().sum())

# Display the first few rows to verify the filtered and cleaned data
dfs[0].head()
```
